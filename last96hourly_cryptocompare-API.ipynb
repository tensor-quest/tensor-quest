```python
import requests
from config import cckey
import pandas as pd
import sys, json
import json

# Get hourly market data for the following cryptocurrencies:

#1. Bitcoin     BTC
#2. Ethereum    ETH
#3. XRP         XRP
#4. Tether      USDT
#5. Cardano     ADA
#6. Monero      XMR
#7. Stellar     XLM
#8. BinanceCoin BNB
#9. Litecoin    LTC
#10.Dogecoin    DOGE

symbols=['BTC','ETH','XRP','USDT','ADA','XMR','XLM','BNB','LTC','DOGE']
for symbol in symbols:

    df = results = dfinput = dfactual = pd.DataFrame()
    url = 'https://min-api.cryptocompare.com/data/v2/histominute?'
    
    # Set the number of datapoints to retrieve
    n = 2000
    
    # Set the number of test sets to generate
    tests = 1
    
    # Set the analysis window size
    WINDOWSIZE = 192
    
    # Initial toTimeStamp needs to be -1 (up to the timestamp of the call)
    # Set totime to zero
    totime = 0
    while (1):
        params = {'fsym':symbol,'tsym':'USD','limit':n,'toTs':totime-1,'api_key':cckey}
        response = requests.get(url, params=params).json()
        if response['Response'] != 'Success':
            break

        # Append the 2000 rows of data to the dataframe
        i = len(response['Data']['Data'])-1
        while (i >= 0):
            data = response['Data']['Data'][i]
            results = pd.DataFrame(data=response['Data']['Data'][i],index=['time'])
            df = df.append(results.set_index('time'))
            i -= 1
            
        # Stop when we reach the nth point
        totime = response['Data']['TimeFrom']
        if len(df) > n:
            break
            
        # Report status
        print("Got to timestamp "+str(totime))
        
    # Report final status and save csv file
    print("Got to timestamp "+str(totime))
    print("Done with "+symbol)
    #df = df[df.index>1642791600]  # Get up to a certain time
    adf = pd.DataFrame()
    for a in range(tests):         # Get the last 10 sets
        adf = df.head(WINDOWSIZE*2+WINDOWSIZE*2*(a+1)).tail(WINDOWSIZE*2)
        adfinput = adf.tail(WINDOWSIZE+1)
        adfactual = adf.head(WINDOWSIZE+1)
        adfinput.to_csv("test"+str(a)+"_"+symbol+".csv")
        adfactual.to_csv("real"+str(a)+"_"+symbol+".csv")
print("Done with everything.")
```
